```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
import re
from ggplot import *

%matplotlib inline
```

```python
pd.read_html('http://www.baseball-reference.com/leagues/MLB/1990-misc.shtml')[0]
```

```python
def fetch_attendance(year):
    url = 'http://www.baseball-reference.com/leagues/MLB/{}-misc.shtml'.format(year)
    data = pd.read_html(url)[0]
    data['year'] = year
    return data
```

```python
fetch_attendance(2013)
```

```python
attendance = pd.concat(fetch_attendance(year) for year in range(1950, 2011))
```

```python
attendance.columns = ["tm", "attendance", "attend_per_game", "batage", "page", "bpf", "ppf", "n_hof",
                      "n_allstars", "n_a_ta_s", "est_payroll", "time", "managers", "year"]
```

```python
attendance.dtypes
```

```python
def make_numeric(num):
    num = str(num)
    num = re.sub(r'[$,]', '', num)
    try:
        return np.float64(num)
    except:
        return np.nan
```

```python
attendance.est_payroll = attendance.est_payroll.map(make_numeric)
```

```python
def fetch_standings(year):
    url = 'http://www.baseball-reference.com/leagues/MLB/{}-standings.shtml'.format(year)
    try:
        data = pd.read_html(url)[1]
    except:
        data = pd.read_html(url)[0]
    data = data.loc[data.Tm != 'Avg']
    data['year'] = year
    return data
```

```python
standings = pd.concat(fetch_standings(year) for year in range(1950, 2011))
```

```python
standings.shape
```

```python
standings.columns = [col.lower() for col in standings.columns]
```

```python
standings.loc[standings.year == 1950]
```

```python
df_merge = pd.merge(attendance, standings, on=['tm', 'year'])
```

```python
df_merge.head()
```

```python
ggplot(aes(x='w', y='attendance'), data=df_merge) + geom_point()
```

```python
ggplot(aes(x='w', y='attend_per_game'), data=df_merge) + geom_point()
```

```python
model = smf.ols("attend_per_game ~ w", df_merge)
```

```python
result = model.fit()
result.summary()
```

```python
model2 = smf.ols("attend_per_game ~ 0 + tm + w", df_merge)
```

```python
result2 = model2.fit()
result2.summary()
```
